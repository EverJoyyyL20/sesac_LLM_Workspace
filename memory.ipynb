```python
# ==============================
# 운영체제(OS) 기능 사용을 위한 기본 라이브러리
# (환경변수 읽기 위해 필요)
# ==============================
import os

# ==============================
# .env 파일을 읽어오기 위한 라이브러리
# (.env 안에 API KEY 같은 비밀 정보 저장)
# ==============================
from dotenv import load_dotenv

# ==============================
# .env 파일 내용을 실제 환경변수로 로드
# ==============================
load_dotenv()

# ==============================
# 환경변수에 저장된 OPENAI_API_KEY 값을 가져오기
# os.getenv("이름") → 환경변수 값 읽기
# ==============================
OPENAI_API_KEY=os.getenv("OPENAI_API_KEY")
```

```python
# ==============================
# LangChain에서 ChatGPT(OpenAI Chat 모델)를 사용하기 위한 클래스 불러오기
# ==============================
from langchain_classic.chat_models import ChatOpenAI

# ==============================
# GPT 모델 객체 생성
#
# temperature=0
# → 답변을 최대한 정확하고 안정적으로 생성
#
# model_name='gpt-4.1-mini'
# → 사용할 GPT 모델 지정 (빠르고 가벼운 모델)
# ==============================
llm=ChatOpenAI(temperature=0,model_name='gpt-4.1-mini')

# ==============================
# LangChain의 대화 체인(ConversationChain) 불러오기
# → 이전 대화 내용을 기억하는 기능 제공
# ==============================
from langchain_classic import ConversationChain


# ==============================
# ConversationChain 객체 생성
#
# llm=llm
# → 위에서 만든 GPT 모델 사용
#
# verbose=True
# → 내부 처리 과정(프롬프트, 메모리 내용) 콘솔에 출력
# ==============================
conversation=ConversationChain(llm=llm,verbose=True)


# ==============================
# 첫 번째 사용자 입력
#
# predict():
# → GPT에게 질문(입력)을 보내는 함수
#
# 이 문장은 메모리에 저장됨
# ==============================
conversation.predict(input="진희는 강아지를 한마리 키우고 있습니다.")

# ==============================
# 두 번째 사용자 입력
#
# 이전 입력과 함께 자동으로 메모리에 누적됨
# ==============================
conversation.predict(input='영수는 고양이를 두마리 키우고 있습니다.')


# ==============================
# 세 번째 질문
#
# 이 질문에는 숫자 정보가 직접 들어있지 않음
# 하지만 ConversationChain이 이전 대화를 기억하고 있어서
#
# 진희: 강아지 1마리
# 영수: 고양이 2마리
#
# → 합계 계산 후 답변 가능
# ==============================
conversation.predict(input="진희와 영수가 키우는 동물은 총 몇마리?")
```
